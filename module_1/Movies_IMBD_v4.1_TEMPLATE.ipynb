# 11. Какого жанра фильмов больше всего?

```python
data = pd.read_csv('movie_bd_v5.xls')
data['genres'] = data['genres'].apply(lambda x:x.split('|'))

import collections

c = collections.Counter()

for category in data['genres']:
    for genre in category:
        c[genre] += 1
        answers['11'] = c
print (answers['11'])
```

# 12. Фильмы какого жанра чаще всего становятся прибыльными? 

```python
data = pd.read_csv('movie_bd_v5.xls')
data['profit'] = data['revenue'] - data['budget'] # нахождение прибыли
data_genres = data.copy()
data_genres = data_genres[data_genres.profit > 0]
answers['12'] = collections.Counter(data_genres.genres.str.cat(sep = '|').split('|')).most_common()
print(answers['12'])
```

# 13. У какого режиссера самые большие суммарные кассовые сбооры?
